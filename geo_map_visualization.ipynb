```python
# geo_map_visualization.ipynb

import pandas as pd
import plotly.express as px

# --- 1. Load your data ---
CSV_PATH = "Cleaned_Viral_Social_Media_Trends.csv"   # adjust path if needed
DATE_COL = "Post_Date"

df = pd.read_csv(CSV_PATH)
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")

# --- 2. Quick check ---
print("Rows:", len(df))
print("Columns:", df.columns.tolist())
df.head()
```

```python
import plotly.express as px

metric = "Engagement_Level"  # or 'Likes', etc.

# Make sure it's numeric
df[metric] = pd.to_numeric(df[metric], errors='coerce')

geo_df = df.groupby("Region", as_index=False)[metric].mean()

fig = px.choropleth(
    geo_df,
    locations="Region",
    locationmode="country names",
    color=metric,
    hover_name="Region",
    color_continuous_scale="Viridis",
    title=f"Average {metric} by Country"
)

fig.update_geos(showcountries=True, projection_type="natural earth")
fig.show()
```
